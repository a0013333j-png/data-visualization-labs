```python
import numpy as np
import matplotlib.pyplot as plt
```

# Programming assignment
Create visualizations for a fictional dataset related to temperature and sales trends over a year. Follow the instructions to produce various types of plots. Ensure your plots are well-labeled and visually clear.


## Step 1: Dataset Creation
1. Create a dataset for months of the year and corresponding data:
* Average monthly temperatures (°C) for your city.
* Monthly sales of ice cream.
* Monthly sales of coffee.
2. Generate synthetic data as follows:
* Temperatures can range between -5°C (winter) to 35°C (summer).
* Ice cream sales should increase with temperature, while coffee sales decrease.


```python
months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]

# 氣溫：設置一個隨月份變化的氣溫範圍，從冬季 (-5°C) 到夏季 (35°C)
temperatures = [-5, 0, 5, 10, 15, 20, 25, 30, 25, 20, 10, 0]

# 根據氣溫生成冰淇淋銷售：假設冰淇淋銷售與氣溫成正比
ice_cream_sales = [temperature * 2 + 10 for temperature in temperatures]

# 根據氣溫生成咖啡銷售：假設咖啡銷售與氣溫成反比
coffee_sales = [35 - temperature * 0.5 for temperature in temperatures]

print("Months:", months)
print("Temperatures:", temperatures)
print("Ice Cream Sales:", ice_cream_sales)
print("Coffee Sales:", coffee_sales)
```

## Step 2: Required Plots


### Line Plot:
Plot average temperature against the months.
Customize the plot with a title, axis labels, and grid lines.


```python
# 假設的數據
months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
temperatures = [-5, 0, 5, 10, 15, 20, 25, 30, 25, 20, 10, 0]

plt.figure(figsize=(8, 5))  # 設置畫布大小
plt.plot(months, temperatures, marker='o', color='b')  # 繪製折線圖
plt.title("Average Temperature Over Months")  # 添加標題
plt.xlabel("Months")  # 添加X軸標籤
plt.ylabel("Temperature (°C)")  # 添加Y軸標籤
plt.grid(True)  # 顯示格線
plt.xticks(rotation=45)  # X軸的月份名稱旋轉45度
plt.show()
```

### Bar Plot:
Create a bar chart showing ice cream and coffee sales side by side for each month.
Use a legend to differentiate between the two.


```python
# Bar Plot - Ice Cream and Coffee Sales
ice_cream_sales = [10, 20, 30, 40, 50, 60, 70, 80, 70, 60, 50, 40]
coffee_sales = [50, 45, 40, 35, 30, 25, 20, 15, 20, 25, 30, 35]

x = np.arange(len(months))  # 月份的索引

fig, ax = plt.subplots(figsize=(10, 6))
bar_width = 0.35  # 設置條形圖的寬度

# 繪製條形圖
ax.bar(x - bar_width/2, ice_cream_sales, bar_width, label='Ice Cream Sales')
ax.bar(x + bar_width/2, coffee_sales, bar_width, label='Coffee Sales')

ax.set_xlabel('Months')
ax.set_ylabel('Sales')
ax.set_title('Ice Cream and Coffee Sales')
ax.set_xticks(x)
ax.set_xticklabels(months, rotation=45)
ax.legend()

plt.show()
```

### Scatter Plot:
Plot a scatter graph to show the relationship between temperature and ice cream sales.
Include axis labels and a trendline (optional).


```python
# Scatter Plot - Temperature vs Ice Cream Sales
plt.figure(figsize=(8, 5))
plt.scatter(temperatures, ice_cream_sales, color='b', label='Ice Cream Sales')  

# 趨勢線
z = np.polyfit(temperatures, ice_cream_sales, 1)  # 擬合一條直線
p = np.poly1d(z)
plt.plot(temperatures, p(temperatures), "orange", label='Trendline')  

plt.title("Temperature vs Ice Cream Sales")
plt.xlabel("Temperature (°C)")
plt.ylabel("Ice Cream Sales")
plt.grid(True)
plt.legend()
plt.show()

```
