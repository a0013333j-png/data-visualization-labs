## Additional exploration of the data
Ask the user to enter two specifc years.  Create a subset of the data including the data for those specifc years. Plot the scatterlots for these two years to compare number of passengers over months in each year

```python
# Get user input
# your code here
plt.figure(figsize=(10,6))
sns.scatterplot(x="year", y="passengers", hue="month", data=flights, palette="Set2")
plt.title("Scatter Plot of Passengers Over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.legend(title="Month")
plt.show()
```

```python
# Validate input
# ensuring that the year is within the acceptable range and creating the filtered data
# Get user input for two specific years
year1 = int(input("Enter the first year: "))
year2 = int(input("Enter the second year: "))
```

```python
# drawing the plot
# Validate input ensuring that the year is within the acceptable range and create the filtered data
if year1 in flights['year'].unique() and year2 in flights['year'].unique():
    filtered_data = flights[flights['year'].isin([year1, year2])]
else:
    print("Invalid year entered. Please enter years within the data range.")
```

```python
# Drawing the scatter plot for the selected years
plt.figure(figsize=(10,6))
sns.scatterplot(x="year", y="passengers", hue="month", data=filtered_data, palette="Set2")
plt.title(f"Comparison of Passengers for {year1} and {year2}")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.legend(title="Month")
plt.show()
```
