```python
# Load an example dataset
# your code goes here
tips = sns.load_dataset("tips")
```

```python
# Create a visualization
# your code goes here
plt.figure(figsize=(8,6))
sns.histplot(data=tips, x="total_bill", hue="time", multiple="stack", kde=False)
plt.title('Stacked Histogram of Bill Amount by Meal Type (Lunch vs Dinner)')
plt.xlabel('Total Bill ($)')
plt.ylabel('Frequency')
plt.show()
```
