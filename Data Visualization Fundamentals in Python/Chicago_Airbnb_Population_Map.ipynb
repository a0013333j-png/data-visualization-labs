```python
#use airbnb data to get the data for the city of chicago, you can access it uner geoda
# Install necessary libraries
!pip install geodatasets
!pip install geopandas
!pip install folium

# Import required libraries
import geodatasets
import geopandas as gpd
import folium
from shapely.geometry import Point, LineString, Polygon

# Use airbnb data to get the data for the city of chicago, you can access it under geoda
chicago = gpd.read_file(geodatasets.get_url("geoda airbnb"))

# Display the first few rows of the dataset to understand its structure
chicago.head()

# Create a Folium map centered on Chicago with an appropriate zoom level
m = folium.Map(location=[41.8781, -87.6298], zoom_start=11)

# Iterate through the Chicago neighborhoods and add a marker for neighborhoods with population larger than 50,000
for idx, row in chicago.iterrows():
    if row['population'] > 50000:  # Only consider neighborhoods with population > 50,000
        # Get the neighborhood name and population for popup
        neighborhood_name = row['community']  # Use 'community' as the neighborhood name
        population = row['population']
        
        # Get the coordinates of the neighborhood (assuming it's a point geometry)
        coords = [row['geometry'].centroid.y, row['geometry'].centroid.x]
        
        # Add a marker with popup for the neighborhood
        folium.Marker(
            location=coords,
            popup=f"{neighborhood_name} - Population: {population}"
        ).add_to(m)

# Display the map
m
```
